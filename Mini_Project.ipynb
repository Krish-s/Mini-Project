```python
df["Ship Mode"] = df["Ship Mode"].apply(lambda i: "First Class" if (i == "Same Day") else i)
df["Ship Mode"] = df["Ship Mode"].apply(lambda i: "Standard Class" if (i == "Not Available") else i)
df["Ship Mode"] = df["Ship Mode"].apply(lambda i: "Second Class" if (i == "unknown") else i)
```

```python
df["Ship Mode"].unique()
```

```python
df["Ship Mode"].fillna(method = "bfill", inplace = True)
```

```python
df.isnull().sum()
```

```python
df["Order Date"]=pd.to_datetime(df["Order Date"],format="%Y-%m-%d")
```

```python
df["Discount"]=df["List Price"]*(df["Discount Percent"]/100)
```

```python
df.head()
```

```python
df.rename(columns={'cost price':'Cost Price'},inplace = True)
```

```python
df.drop(columns="discount",axis=1,inplace= True)
```

```python
df["Sale Price"] = df["List Price"]-df["Discount"]
```

```python
df.head()
```

```python
df["Profit"]=df["Sale Price"]-df["Cost Price"]
```
