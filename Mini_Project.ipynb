```python
--#My Queries
#--1.Highest profit products
select sub_category,profit from table1 t1 join table2 t2 on t1.order_id = t2.order_id
group by sub_category,profit order by profit desc;


#--2.Total Profit by region
select region ,sum(profit) as total_profit from table1 t1 join table2 t2 on t1.order_id = t2.order_id
group by region order by total_profit;

#--3.Lowest sale price per each sub category
select sub_category, min(sale_price) as min_sale_price from table1 t1 join table2 t2 on t1.order_id = t2.order_id
group by sub_category order by min_sale_price;

#--4.create a database orders,create 2 tables in that database
create database orders;
create table table1();
create table table2();

#--5.Order,ship mode, city with discount percent maximum
select ship_mode,city,max(discount_percent) as max_discount from table1 t1 join table2 t2 on t1.order_id = t2.order_id
 group by ship_mode,city order by max_discount desc;


#--6.Counts of order by shipping mode
select ship_mode,count(ship_mode) as total_ship_mode from table1 t1 join table2 t2 on t1.order_id = t2.order_id
group by ship_mode order by total_ship_mode desc;

#--7.Total sales by state
select state,sum(sale_price) as total_sales from table1 t1 join table2 t2 on t1.order_id = t2.order_id
group by state order by total_sales desc;

#--8.Most frequently ordered product
select product_id,count(*) as order_count from table1 t1 join table2 t2 on t1.order_id = t2.order_id
group by product_id order by order_count desc;

#--9.Orders with sale price less than cost price
select * from table1 t1 join table2 t2 on t1.order_id = t2.order_id
where sale_price < cost_price;

#--10.Most frequently ordering cities
select city,count(*) as total_orders from table1 t1 join table2 t2 on t1.order_id = t2.order_id
group by city order by total_orders desc;

##--11.To print order id , product id , subcategory , total sales
select t1.order_id as t1_order_id,product_id,sub_category,sum(sale_price) as total_sale
from table1 as t1 join table2 as t2 on t1.order_id = t2.order_id group by t1.order_id,product_id,sub_category
order by total_sale desc;
```
